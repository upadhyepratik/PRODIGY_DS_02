```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d1 = pd.read_csv('/content/drive/MyDrive/PRODIGY_DATA_SCIENCE_INTERNSHIP/train.csv')
d1.head()
```

```python
d2 = pd.read_csv('/content/drive/MyDrive/PRODIGY_DATA_SCIENCE_INTERNSHIP/test.csv')
d2.head()
```

```python
d3 = pd.read_csv('/content/drive/MyDrive/PRODIGY_DATA_SCIENCE_INTERNSHIP/gender_submission.csv')
d3.head()
```

```python
d1.isnull().sum()
```

```python
d2.isnull().sum()
```

```python
d3.isnull().sum()
```

```python
d1['Age'].fillna(d1['Age'].median(), inplace=True)
d1
```

```python
d2['Age'].fillna(d2['Age'].median(), inplace=True)
d2
```

```python
d1.drop(['Cabin','PassengerId','Name','Ticket'], axis=1, inplace=True)
d1
```

```python
d2.drop(['Cabin','PassengerId','Name','Ticket'], axis=1, inplace=True)
d2
```

```python
d1['Embarked'] = d1['Embarked'].fillna(method = 'ffill')
d1
```

```python
d1['Embarked'] = d1['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
d1
```

```python
d2['Embarked'] = d2['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
d2
```

```python
d1['Sex'] = d1['Sex'].map({'male': 0, 'female': 1})
d1
```

```python
d2['Sex'] = d2['Sex'].map({'male': 0, 'female': 1})
d2
```

# **Exploratory Data Analysis (EDA)**

```python
d1.describe()
```

```python
d2.describe()
```

```python
d3.describe()
```

```python
sns.pairplot(d1, hue='Survived')
plt.show()
```

```python
sns.pairplot(d2)
plt.show()
```

```python
#correlation matrix of train data
cm = d1.corr()
sns.heatmap(cm, annot=True, cmap='viridis')
plt.show()
```

```python
#correlation matrix of test data
cm2 = d2.corr()
sns.heatmap(cm2, annot=True, cmap='viridis')
plt.show()
```

```python
#age distribution among survivors and non-survivors
plt.figure(figsize=(10, 5))
sns.histplot(d1[d1['Survived'] == 1]['Age'], label='Survived', kde=True)
sns.histplot(d1[d1['Survived'] == 0]['Age'], label='Did not survive', kde=True)
plt.legend()
plt.show()
```

```python
#fare distribution in test
plt.figure(figsize=(10, 3))
sns.histplot(d1['Fare'], kde=True)
plt.title('Fare Distribution in Test Data')
plt.show()
```

```python
#'Survived' distribution based on gender
sns.countplot(data=d3, x='Survived')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=d1)
plt.show()
#0=male
#1=female
```

```python
#Age Distribution
sns.histplot(x='Age', data=d1, kde=True)
plt.show()
```
